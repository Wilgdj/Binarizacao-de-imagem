# Imagem tons de cinza

```python
from PIL import Image

image_path = "/content/image/sunset.jpg"
image = Image.open(image_path)
image
```

```python
pixels = image.load()
width, height = image.size
```

```python
gray_image = Image.new("L", (width, height))
gray_pixels = gray_image.load()
```

```python
for y in range(height):
  for x in range(width):

    r, g, b = pixels[x, y]
    gray = int(0.3 * r + 0.59 * g + 0.11 * b)
    gray_pixels[x, y] = gray
```

```python
gray_image.save("imagem_cinza.jpg")
gray_image.show()
```

```python
image_gray_path = "/content/imagem_cinza.jpg"
image = Image.open(image_gray_path)
image
```

#Imagem preta e branca

```python
from PIL import Image

image_path = "/content/image/sunset.jpg"
image = Image.open(image_path)
image
```

```python
gray_image = image.convert("L")
threshold = 128
bw_image = gray_image.point(lambda x: 255 if x > threshold else 0, mode="1")
```

```python
bw_image.save("imagem_bw.jpg")
bw_image.show()
```

```python
image_bw_path = "/content/imagem_bw.jpg"
image = Image.open(image_bw_path)
image
```
